```python
from transformers import pipeline
from langchain_community.document_loaders import PyPDFLoader
from newspaper import Article
import re

# Load a summarization-specific model
print("Creating pipeline...")
summarizer = pipeline("summarization", model="facebook/bart-large-cnn", device=-1)  # Use CPU

# Load a question-answering model
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")

def truncate_text(text, max_tokens=1024):
    # Split the text into tokens (words) and truncate if necessary
    tokens = text.split()
    if len(tokens) > max_tokens:
        tokens = tokens[:max_tokens]
        text = " ".join(tokens)
    return text

def clean_text(text):
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    # Remove special characters and extra spaces
    text = re.sub(r'\s+', ' ', text)  # Replace multiple spaces with a single space
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)  # Remove non-ASCII characters
    return text.strip()

def summarize_text(text):
    # Clean and truncate the text
    text = clean_text(text)
    text = truncate_text(text)

    # Generate the summary
    print("Generating summary...")
    summary = summarizer(
        text,
        max_length=200,  # Increase max_length
        min_length=50,   # Increase min_length
        max_new_tokens=150,  # Increase max_new_tokens
        do_sample=False,
    )
    return summary[0]["summary_text"]

def load_pdf(file_path):
    # Use PyPDFLoader to load the PDF
    loader = PyPDFLoader(file_path)
    documents = loader.load()
    # Combine all pages into a single text
    text = " ".join([doc.page_content for doc in documents])
    return text

def fetch_article_from_url(url):
    """
    Extrait le contenu d'un article à partir de son URL.
    """
    # Créer un objet Article
    article = Article(url)

    # Télécharger et parser l'article
    article.download()
    article.parse()

    # Retourner le texte de l'article
    return article.text

def ask_question(article, question):
    """
    Answer a question based on the article text.
    """
    try:
        # Use the QA pipeline to get the answer
        result = qa_pipeline(question=question, context=article)
        return result["answer"]
    except Exception as e:
        return f"Error answering the question: {e}"

def main():
    print("Choose an option:")
    print("1. Paste the article")
    print("2. Load the article from a file")
    print("3. Enter the URL of an article")
    choice = input("Enter your choice (1, 2, or 3): ")

    if choice == "1":
        # User pastes the article
        article = input("Paste your article here:\n")
    elif choice == "2":
        # User loads the article from a file
        file_path = input("Enter the path to the file: ")
        if file_path.endswith(".pdf"):
            # Extract text from PDF using PyPDFLoader
            try:
                article = load_pdf(file_path)
            except Exception as e:
                print(f"Error reading PDF file: {e}")
                return
        else:
            # Read plain text file
            try:
                with open(file_path, "r", encoding="utf-8") as file:
                    article = file.read()
            except Exception as e:
                print(f"Error reading file: {e}")
                return
    elif choice == "3":
        # User enters the URL of an article
        url = input("Enter the URL of the article: ")
        try:
            article = fetch_article_from_url(url)
        except Exception as e:
            print(f"Error fetching article from URL: {e}")
            return
    else:
        print("Invalid choice. Please try again.")
        return

    # Summarize the article
    summary = summarize_text(article)
    print("\nGenerated Summary:")
    print(summary)

    # Allow the user to ask questions
    while True:
        question = input("\nAsk a question about the article (or type 'exit' to quit): ")
        if question.lower() == "exit":
            break
        answer = ask_question(article, question)
        print(f"Answer: {answer}")

if __name__ == "__main__":
    main()
```
